```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as ply
%matplotlib inline
```

```python
# Import data and convert to csv, then print

data = pd.read_excel('contributions.xlsx', index_col=0)
data.to_csv('contributions.csv', encoding='utf-8')
data
```

```python
# Display all columns, then print

pd.options.display.max_columns = None
data
```

```python
# Clean data, then print

data.columns = data.columns.str.strip().str.replace('.', '') # Remove whitespace, period chars (to later delete useless column)
data = data.replace(np.NaN, 0) # Replace instances of 0 with NaN
data = data.drop(['Doc No', 'Career Background', 'Country Experience', 'Regional Experience', 'Language'], axis=1) # Remove useless columns

data
```

```python
# Convert columns from data type object to float64, then print

data['Total Personal Contributions by Nominee'] = pd.to_numeric(data['Total Personal Contributions by Nominee'], errors='coerce')
data['Bundling Value'] = pd.to_numeric(data['Bundling Value'], errors='coerce')

data
```

```python
# Grouping exercise

data[data['Experience in Organizational Leadership'] == 'Yes'].groupby('President')['Total Personal Contributions by Nominee'].sum()
```

```python
# Grouping exercise

data[data['Graduate Degree(s) & Area(s) of Study'] != 'No'].groupby('President')['Total Personal Contributions by Nominee'].sum()
# data[data['Total Personal Contributions by Nominee'] == 'call'].groupby(['network', 'month'])['duration'].sum()
```

```python
# Basic graph indicating 2005 was big contribution year

data.plot(y='Total Personal Contributions by Nominee', x='Nomination Year')
```

```python
# Pie chart of contribution sums

pres = ['Bush', 'Clinton', 'Obama', 'Reagan', 'Trump']

# Calculate sum of pres contributions
pres_sum = data[data['Total Personal Contributions by Nominee'] != 'NaN'].groupby('President')['Total Personal Contributions by Nominee'].sum()

fig1, ax1 = ply.subplots()
ax1.pie(pres_sum,
        labels=pres,
        autopct='%1.1f%%',
        shadow=True,
        startangle=90)
ax1.axis('equal')

ply.show()
```
