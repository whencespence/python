```python
#Function to convert val to better month format

def get_month_name(val):
    months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sept','Oct','Nov','Dec']
    
    #error checking
    try:
        return months[val - 1]
    except:
        return val
```

```python
data['Month'] = pd.Datetime(data['Occurrence_year'])
data
```

```python
data['Month'] = data['Month'].apply(get_month_name)
```
